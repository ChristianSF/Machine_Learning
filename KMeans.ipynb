```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

%matplotlib inline
```

```
dados = pd.read_csv("iris.csv")
```

```
dados.head()
```

```
sb.pairplot(dados, hue="variety")
```

```
sb.pairplot(dados)
```

```
X = np.array(dados.drop('variety', axis = 1))
```

```
X
```

```
from sklearn.cluster import KMeans
```

```
kmeans = KMeans(n_clusters=3, random_state=0)
```

```
kmeans.fit(X)
```

```
kmeans.labels_
```

```
dados['K-classes'] = kmeans.labels_
```

```
dados
```

```
sb.pairplot(dados, hue='variety')
```

```
sb.pairplot(dados, hue='K-classes')
```
